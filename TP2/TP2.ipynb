### Exercice 1
Exercice 1 -Utilisation de NumPy
Les nombres aléatoires
1. Générer un numpy array de 10 valeurs aléatoires.
2. Générer un numpy array de 100 valeurs entières entre 0 et 100.
3. Générer un numpy array de taille 25 avec des nombres tirés aléatoirement suivant une distribution normale centrée réduite.
4. Générer un numpy array de taille 25 avec des nombres tirés aléatoirement suivant une loi uniforme.
5. Générer une matrice aléatoire de dimension 2 et de taille 10×10.
6. Modifier votre programme pour que les nombres aléatoires générés soient toujours les mêmes à chaque exécution.

```python
import numpy as np
np.random.seed(0)

random10 = np.random.random(10)
print("Numpy array de 10 valeurs aléatoires : ")
print(str(random10)+"\n")

random100 = np.random.randint(0,101,100)
print("Numpy array de 100 valeurs entières aléatoires : ")
print(str(random100)+"\n")

random_norm = np.random.standard_normal(25)
print("Numpy array de 25 valeurs tirés suivant une distribution normale centrée réduite : ")
print(str(random_norm)+"\n")

random_unif = np.random.uniform(0,1,25)
print("Numpy array de 25 valeurs tirés suivant une distribution uniforme dans [0,1] : ")
print(str(random_unif)+"\n")

matrandom = np.random.random((10,10))
print("Matrice aléatoire 10x10 : ")
print(str(matrandom)+"\n")
```

Les entrées/sorties
1. Sauvegarder une matrice aléatoire de taille 10 × 10 dans un fichier matrice1.csv avec ’ ;’ comme séparateur.
2. Charger le fichier matrice1.csv.

```python
matrice1 = np.random.random((10,10))
print("Matrice avant écriture : ")
print(str(matrice1)+"\n")
np.savetxt("matrice1.csv",matrice1,delimiter=";")
matrice2 = np.loadtxt("matrice1.csv",delimiter=";")
print("Matrice après lecture : ")
print(str(matrice2)+"\n")
```

### Exercice 2 - Utilisation de pandas
1. Construire un dataframe de quatre colonnes nommées A, B, C et D et trois lignes numérotées i1, i2 et i3 remplies de 1.
2. Construire un dataframe de quatre colonnes nommées p1, p2, p3 et p4 et 100 lignes numérotées de i1 à i100 remplies du
chiffre 1 uniquement pour l’instant.

```python
import pandas as pd

df1 = pd.DataFrame(data = np.ones((3,4)),columns = ["A","B","C","D"],index = ["i1","i2","i3"])
print("Data Frame n°1 :")
print(df1)

data = np.ones((100,4))
col = ["p1","p2","p3","p4"]
idx = ["i"+str(i) for i in range(1,101)]
df2 = pd.DataFrame(data=data,columns=col,index=idx)
print("Data Frame 2 :")
print(df2)
```

3. Modifier le dataframe pour que la colonne p1 contiennent tous les nombres de 1 à 100 en ligne à la puissance 1, jusqu’à p4 contenant tous les nombres de 1 à 100 à la puissance 4 en ligne.

```python
df2["p2"]=np.arange(1,101)**2
df2["p3"]=np.arange(1,101)**3
df2["p4"]=np.arange(1,101)**4
print("Data Frame 2 après modif : ")
print(df2)
```

4. Afficher la forme du dataframe.
5. Afficher les 10 premiers éléments du dataframe.
6. Afficher les 10 derniers éléments du dataframe.
7. Afficher le noms des colonnes.
8. Afficher le type de données des colonnes.
9. Afficher les statistiques pour le dataframe.
10. Afficher les données de la colonne p4 uniquement.
11. Afficher les données des colonnes p1 et p4 uniquement.
12. Afficher la moyenne des lignes et la moyenne des colonnes de ce dataframe.

```python
print("La forme de df2 est :" + str(df2.shape))
print("Les 10 premiers éléments de df2 sont : \n"+str(df2.head(10))+"\n")
print("Les 10 derniers éléments de df2 sont : \n"+str(df2.tail(10))+"\n")
# L'appel à la méthode tolist() n'est utile que pour avoir un meilleur affichage
print("Les colonnes sont : \n" +str(df2.columns.tolist())+"\n")
print("Le type de données des colonnes : \n"+str(df2.dtypes.tolist())+"\n")
print("Les statistiques du dataframe : \n"+str(df2.describe(include="all"))+"\n")
print("Les données de p4 sont : \n"+str(df2["p4"])+"\n")
print("Les données de p1 et p4 sont : \n"+str(df2[['p1', 'p4']])+"\n")
print("La moyenne des lignes est : \n"+str(df2.mean(axis=1))+"\n")
print("La moyenne des colonnes est : \n"+str(df2.mean(axis=0))+"\n")
```
