### Exercice 1 
1. Construire une liste liste1 qui contient les nombres de 1 à 99 inclus (avec la fonction range).
2. Ajouter à l’aide d’une boucle les éléments de 100 à 199, à une liste vide liste2.
3. En une opération, ajouter tous les éléments de la liste liste2 à la liste liste1.

```python
liste1 = [i for i in range(1,100)]
print("Liste 1 : \n" + str(liste1) + "\n")
liste2 = [i for i in range(100,200)]
print("Liste 2 : \n" + str(liste2) + "\n")
liste1 += liste2
print("Liste 1 après concaténation de liste 2 \n" + str(liste1)+"\n")
```

4. Avec la fonction range(), construire une liste3 composée de tous les nombres pairs de 0 à 100
(inclus).
5. À l’aide d’une boucle supprimez tous les multiples de 10 contenus dans la liste liste3.
6. Affichez la longueur de la liste liste3.
7. Affichez tous les multiples de 8 de liste3.
8. Affichez les 20 premiers éléments de la liste liste3 (sans faire de boucle).
9. Affichez les éléments suivants les 20 premiers éléments de liste3.
10. Affichez les 10 derniers éléments de liste3.

```python
liste3 = [i for i in range(0, 101, 2)]
print("Liste 3 (nombres pairs de 0 à 100) : \n" + str(liste3) + "\n")

liste3 = [i for i in liste3 if i % 10 != 0]
print("Liste 3 après suppression des multiples de 10 : \n" + str(liste3) + "\n")

print("Longueur de liste 3 : " + str(len(liste3)) + "\n")

print("Les multiples de 8 dans liste 3 sont :")
print(str([i for i in liste3 if i % 8 == 0]) + "\n")

print("Les 20 premiers termes de liste 3 sont :")
print(str(liste3[:20])+"\n")

print("Les éléments suivants les 20 premiers éléments de liste3 sont :")
print(str(liste3[20:])+"\n")
print("Les 10 derniers éléments de liste 3 sont :")
print(str(liste3[-10:])+"\n")
```

14. Créer une liste6 avec les valeurs [1,2,3] puis créer une copie de liste6 dans une liste6bis
(attention si on change une valeur de liste6bis, la liste6 ne doit pas être modifiée). cf.

```python
liste4 = [i for i in range(0, 51) if i % 2 == 0]
print("Liste 4 (nombres pairs de 0 à 50) : \n" + str(liste4) + "\n")

listWords = ["hello", "the", "world", "apple", "tree", "lemon"]
listWords5 = [w for w in listWords if len(w) < 5]
print("Liste des mots de moins de 5 caractères : \n" + str(listWords5) + "\n")

liste5 = [0, 1]
for k in range(2, 20):
    liste5.append(liste5[k - 1] + liste5[k - 2])
print("Liste Fibonacci jusqu'au vingtième terme : \n" + str(liste5) + "\n")

liste6 = [1, 2, 3]
liste6bis = liste6.copy()
liste6bis.append(1)
print("Liste 6 originale : \n" + str(liste6) + "\n")
print("Copie de liste 6 après ajout d'un élément dans celle-ci : \n" + str(liste6bis) + "\n")
```
